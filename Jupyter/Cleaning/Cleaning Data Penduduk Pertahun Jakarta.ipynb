```python
import pandas
import numpy
```

```python
data_jumlah_penduduk_17_19 = pandas.read_excel("../../Dataset/Banjir/Jumlah Penduduk Menurut Kabupaten_Kota di Provinsi DKI Jakarta.xlsx")
data_jumlah_penduduk_20_22 =  pandas.read_excel("../../Dataset/Banjir/Jumlah Penduduk Menurut Kabupaten_Kota di Provinsi DKI Jakarta_20_22.xlsx")
```

```python
data_jumlah_penduduk_17_19
```

```python
column = ["Kabupaten/Kota","2017","2018","2019","2020","2021","2022"]
data_jumlah_penduduk = pandas.DataFrame(columns=column)
```

```python
for i in range(1,8):
    data_jumlah_penduduk = data_jumlah_penduduk.append({
        "Kabupaten/Kota":data_jumlah_penduduk_17_19["Kab/Kota"][i],
        "2017":data_jumlah_penduduk_17_19["Jumlah Penduduk Menurut Kabupaten/Kota di Provinsi DKI Jakarta (Jiwa)"][i],
        "2018":data_jumlah_penduduk_17_19["Unnamed: 2"][i],
        "2019":data_jumlah_penduduk_17_19["Unnamed: 3"][i],
        "2020":data_jumlah_penduduk_20_22["Jumlah Penduduk Menurut Kabupaten/Kota di Provinsi DKI Jakarta (Jiwa)"][i],
        "2021":data_jumlah_penduduk_20_22["Unnamed: 2"][i],
        "2022":data_jumlah_penduduk_20_22["Unnamed: 3"][i]
    }, ignore_index=True)
```

```python
data_jumlah_penduduk.info()
```

```python
data_jumlah_penduduk["2017"] = data_jumlah_penduduk["2017"].astype(int)
data_jumlah_penduduk["2018"] = data_jumlah_penduduk["2018"].astype(int)
data_jumlah_penduduk["2019"] = data_jumlah_penduduk["2019"].astype(int)
data_jumlah_penduduk["2020"] = data_jumlah_penduduk["2020"].astype(int)
data_jumlah_penduduk["2021"] = data_jumlah_penduduk["2021"].astype(int)
data_jumlah_penduduk["2022"] = data_jumlah_penduduk["2022"].astype(int)
```
