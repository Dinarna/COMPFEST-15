```python
def cekkosong(data,kolom,tahun):
    print(tahun)
    print(data.isnull().any())
    print()
    print(data.isnull().sum())
    print("-"*50)
    print()
    return 
```

```python
def misahin_rata_rata_ketinggian_air(data,kolom):
    himpunan_nilai_baru=[]
    for i in data[kolom]:
        if isinstance(i, str) == True:
            replacements = {
                's.d': '-',
                'n.a':'0',
                's/d': '-',
                'cm': '',
                'May': '0',
                'Oct': '0',
                ' ':''
                }
            pola = re.compile('|'.join(re.escape(key) for key in replacements.keys()))
            output = pola.sub(lambda match: replacements[match.group(0)], i)
        himpunan_nilai_baru.append(output)
    data[kolom]=himpunan_nilai_baru
    data[['rata_rata_ketinggian_air_min', 'rata_rata_ketinggian_air_max']] = data[kolom].str.split('-', expand=True)
    return data
```

```python
def gantibulandantahun(data,kolom,tahun):
    himpunan_bt=[]
    for bulan in data[kolom]:
        jadi_string = str(bulan)
        bulan_tahun=jadi_string+str(tahun)
        himpunan_bt.append(bulan_tahun)
    data[kolom] = himpunan_bt
    data[kolom] = pandas.to_datetime(data[kolom],format='%m%Y', errors='coerce').dt.to_period('m')
    return  
```

```python
tahun_data=2013
for tahun in himpunan_tahun :
    gantibulandantahun(tahun,"bulan",tahun_data)
    tahun_data+=1
rekap_banjir_2013_2020 = pandas.concat(himpunan_tahun, ignore_index=True)
```
